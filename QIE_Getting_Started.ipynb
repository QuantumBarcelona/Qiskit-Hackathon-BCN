```python
from QIE import *
```

```python
image = generate_random_image()
```

# Version 1
- Signal intensity encoded as amplitude of each state
- Max amplitude stored in a classical register

```python
depth_v1 = QIE_V1(image, get_depth=True)
print('Depth:', depth_v1)
```

# Version 1.1
- Signal intensity encoded as amplitude of each state
- Max amplitude stored in an extra qubit

```python
depth_v1_1 = QIE_V1_1(image, get_depth=True)
print('Depth:', depth_v1_1)
```

# Version 2
- Ancillary Qubit used as target qubit for each combination of high filter / low filter control gate

```python
depth_v2 = QIE_V2(image, get_depth=True)
print('Depth:', depth_v2)
```

# Version 3
- Clustering of pixels by K-Means (8x8 to 4x4)
- Signal intensity encoded as the amplitude of each state

```python
depth_v3 = QIE_V3(image, get_depth=True)
print('Depth:', depth_v3)
```

# Version 3.1
- Clustering of pixels by K-Means (16x16 to 8x8)
- Signal intensity encoded as the amplitude of each state

```python
image_16 = generate_random_image(16, 16)

depth_v3_1 = QIE_V3_1(image_16, get_depth=True)
print('Depth:', depth_v3_1)
```

# Version 4
- Clustering into buckets of 2x2 pixels, but staying 8x8
- Amplitude is averaged in each bucket
- Every pixel can be separately targeted in the quantum state

```python
depth_v4 = QIE_V4(image, get_depth=True)
print('Depth:', depth_v4)
```
